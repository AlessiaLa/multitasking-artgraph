```python
from load_dataset.artgraph import ArtGraph
import pandas as pd
import numpy as np
```

```python
root = 'artgraph2bestemotions'
```

```python
data = ArtGraph(root = root, preprocess='constant', features = 'vit', fine_tuning = True)[0]
data
```

```python
emotion2artwork = pd.DataFrame(data['artwork', 'emotion'].edge_index.T.numpy(), columns = ['artwork', 'emotion'])
emotion2artwork
```

```python
style2artwork = pd.DataFrame(data['artwork', 'style'].edge_index.T.numpy(), columns = ['artwork', 'style'])
style2artwork
```

```python
genre2artwork = pd.DataFrame(data['artwork', 'genre'].edge_index.T.numpy(), columns = ['artwork', 'genre'])
genre2artwork
```

```python
data_labels = pd.merge(pd.merge(emotion2artwork, style2artwork, on = 'artwork'), genre2artwork, on = 'artwork')
```

```python
data_labels
```

```python
data_labels[['artwork', 'style', 'genre', 'emotion']].to_csv('data_full_labels.csv', index = False)
```
